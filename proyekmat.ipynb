```python
# Menghitung Hampiran Daya
# Integrasi Parabola, Trapesium, dan Riemann Kiri

'''
v(t) = 230√2 sin(100pi t) 
i(t) = 5 sin(100pi t − pi/6)
p(t) = v(t) × i(t)
T = 1/50
'''

import math

a = 0 

b = T = 1/50

n = int(input("Masukkan partisi = "))
if n % 2 != 0:
    print("Input harus bilangan genap.")
    n = int(input("Masukan partisi = "))

def p(t):
    return 230*math.sqrt(2)*math.sin(100*math.pi*t)*5*math.sin(100*math.pi*t-math.pi/6)
```

```python
print("Mari menghitung dengan metode integrasi numerik parabola!")

def simpsons (p,a,b,n):   
    h = (b-a)/n 
    x=[] #koordinat x

    for i in range (n+1): #0,1,2,3,..,n
        x.append(a)
        a += h
    
    ganjil=0
    genap=0
    
    for i in range (1,n): #1,2,...,n-1
        if i%2==0:
            genap += p(x[i])
        else:
            ganjil += p(x[i])

    return (h/3)*((p(a) + 4*ganjil+ 2*genap + p(b)))

hasil = simpsons(p, a, b, n)
daya = hasil*(1/(b-a))
print(f"Hasil integrasi numerik metode parabola yaitu {hasil:18f}")
print(f"Daya rata-rata diperkirakan yaitu {daya:9f} Watt")
```

```python
print("Mari menghitung dengan metode integrasi numerik trapesium!")

def trap (p,a,b,n):

    deltat=(b-a)/n

    hasil=0
    for i in range(1,n):
        ti=a+(i*deltat)
        hasil=hasil+(2*p(ti))

    hasil=hasil+p(a)+p(b)

    hasil=hasil*deltat/2

    return hasil

hasil=trap(p,a,b,n)
daya=hasil*(1/(b-a))
print(f"Hasil integrasi numerik metode trapesium yaitu {hasil:18f}")
print(f"Daya rata-rata diperkirakan yaitu {daya:9f} Watt")
```

```python
print("Mari menghitung dengan metode integrasi numerik persegi panjang kiri!")

def riemann_kiri(p,a,b,n):
    h = (b-a)/n

    hasil = 0
    for k in range (1,n+1):
        hasil += p(a+((n-0.5)*h))
        
    hasil += hasil*h

    return hasil

hasil=riemann_kiri(p,a,b,n)
daya=hasil*(1/(b-a))
print(f"Hasil integrasi numerik metode persegi panjang kiri yaitu {hasil:18f}")
print(f"Daya rata-rata diperkirakan yaitu {daya:9f} Watt")
```
