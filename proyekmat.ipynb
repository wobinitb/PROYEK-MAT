```python
# Menghitung Hampiran Daya
# Integrasi Parabola, Trapesium, dan Riemann Kiri
import math
a = 0 
b = T = 1/50
Vm = 230*math.sqrt(2)
Im = 5
frekuensi = 50
Kecepatan_angular = 2*math.pi*frekuensi
Sudut = math.pi/6
Periode = 1/frekuensi

n = int(input("Masukkan partisi = "))
if n % 2 != 0:
    print("Input harus bilangan genap.")
    n = int(input("Masukan partisi = "))

def v(t):
    return Vm*math.sin(Kecepatan_angular*t)
def i(t):
    return Im*math.sin(Kecepatan_angular*t-Sudut)
def p(t):
    return v(t)*i(t)
```

```python
print("Mari menghitung dengan metode integrasi numerik parabola!")

def simpsons (p,a,b,n):   
    h = (b-a)/n 
    x=[a + i*h for i in range(n+1)]
    ganjil=0
    genap=0
    for i in range (1,n): #1,2,...,n-1
        if i%2==0:
            genap += p(x[i])
        else:
            ganjil += p(x[i])
    return (h/3)*((p(a) + 4*ganjil+ 2*genap + p(b)))

hasil = simpsons(p, a, b, n)
daya = hasil/(b-a)
print(f"Daya rata-rata diperkirakan yaitu {daya} Watt")
```

```python
print("Mari menghitung dengan metode integrasi numerik trapesium!")

def trap (p,a,b,n):
    deltat=(b-a)/n
    hasil=0
    for i in range(1,n):
        ti=a+(i*deltat)
        hasil=hasil+(2*p(ti))
    hasil=hasil+p(a)+p(b)
    hasil=hasil*deltat/2*(1/(b-a))
    return hasil

daya=trap(p,a,b,n)
print(f"Daya rata-rata diperkirakan yaitu {daya} Watt")
```

```python
print("Mari menghitung dengan metode integrasi numerik persegi panjang kiri!")

def riemann_kiri(p,a,b,n):
    h = (b-a)/n
    hasil = 0
    for i in range (0,n):
        ti=a+(i*h)
        hasil+=(p(ti))*h
    return hasil/(b-a)

daya=riemann_kiri(p,a,b,n)
print(f"Daya rata-rata diperkirakan yaitu {daya} Watt")
```
