```python
# kode numerik trapesium
import math
import numpy as np
import matplotlib.pyplot as plt
a = 0 
b = T = 1/50
Vm = 230*math.sqrt(2)
Im = 5
frekuensi = 50
Kecepatan_angular = 2*math.pi*frekuensi
Sudut = math.pi/6
Periode = 1/frekuensi

n = int(input("Masukkan partisi = "))
if n % 2 != 0:
    print("Input harus bilangan genap.")
    n = int(input("Masukan partisi = "))

def v(t):
    return Vm*np.sin(Kecepatan_angular*t)
def i(t):
    return Im*np.sin(Kecepatan_angular*t-Sudut)
def p(t):
    return v(t)*i(t)

print("Mari menghitung dengan metode integrasi numerik trapesium!")

def trap (p,a,b,n):
    deltat=(b-a)/n
    hasil=0
    for i in range(1,n):
        ti=a+(i*deltat)
        hasil=hasil+(2*p(ti))
    hasil=hasil+p(a)+p(b)
    hasil=hasil*deltat/2*(1/(b-a))
    return hasil

daya=trap(p,a,b,n)
print(f"Daya rata-rata diperkirakan yaitu {daya} Watt")
```

```python
# kode numerik riemann kiri
import math
import numpy as np
import matplotlib.pyplot as plt
a = 0 
b = T = 1/50
Vm = 230*math.sqrt(2)
Im = 5
frekuensi = 50
Kecepatan_angular = 2*math.pi*frekuensi
Sudut = math.pi/6
Periode = 1/frekuensi

n = int(input("Masukkan partisi = "))
if n % 2 != 0:
    print("Input harus bilangan genap.")
    n = int(input("Masukan partisi = "))

def v(t):
    return Vm*np.sin(Kecepatan_angular*t)
def i(t):
    return Im*np.sin(Kecepatan_angular*t-Sudut)
def p(t):
    return v(t)*i(t)

print("Mari menghitung dengan metode integrasi numerik persegi panjang kiri!")

def riemann_kiri(p,a,b,n):
    h = (b-a)/n
    hasil = 0
    for i in range (0,n):
        ti=a+(i*h)
        hasil+=(p(ti))*h
    return hasil/(b-a)

daya=riemann_kiri(p,a,b,n)
print(f"Daya rata-rata diperkirakan yaitu {daya} Watt")
```

```python
#Membandingkan 3 metode dengan 3 variasi n
import math
import numpy as np
import matplotlib.pyplot as plt
a = 0 
b = T = 1/50
Vm = 230*math.sqrt(2)
Im = 5
frekuensi = 50
Kecepatan_angular = 2*math.pi*frekuensi
Sudut = math.pi/6
Periode = 1/frekuensi

n_list = [6, 500, 10000]
hasil_simpsons = []
hasil_trap = []
hasil_riemann = []

def v(t):
    return Vm*np.sin(Kecepatan_angular*t)
def i(t):
    return Im*np.sin(Kecepatan_angular*t-Sudut)
def p(t):
    return v(t)*i(t)

def simpsons (p,a,b,n):   
    h = (b-a)/n
    x=np.linspace(a, b, n+1)
    ganjil=0
    genap=0
    for i in range (1,n): #1,2,...,n-1
        if i%2==0:
            genap += p(x[i])
        else:
            ganjil += p(x[i])
    return ((h/3)*((p(a) + 4*ganjil+ 2*genap + p(b))))/(b-a)

def trap (p,a,b,n):
    h = (b-a)/n
    hasil=0
    for i in range(1,n):
        ti=a+(i*h)
        hasil=hasil+(2*p(ti))
    hasil=hasil+p(a)+p(b)
    hasil=hasil*h/2*(1/(b-a))
    return hasil

def riemann_kiri(p,a,b,n):
    h = (b-a)/n
    hasil = 0
    for i in range (0,n):
        ti=a+(i*h)
        hasil+=(p(ti))*h
    return hasil/(b-a)

for n in n_list:
    hasil_simpsons.append(simpsons (p,a,b,n))
    hasil_trap.append(trap (p,a,b,n))
    hasil_riemann.append(riemann_kiri (p,a,b,n))

print('Perbandingan hasil ketiga cara dengan partisi beragam')
print("=" * 80)
print(f"{'n':<8} {'Parabola':<20} {'Trapezoidal':<20} {'Riemann Kiri':<20}")
print("-" * 80)
for i in range(len(n_list)):
    print(f"{n_list[i]:<8} {hasil_simpsons[i]:<20.15f} {hasil_trap[i]:<20.15f} {hasil_riemann[i]:<20.15f}")
print("=" * 80)
```

```python
#grafik fungsi P I V
import math
import numpy as np
import matplotlib.pyplot as plt
a = 0 
b = T = 1/50
Vm = 230*math.sqrt(2)
Im = 5
frekuensi = 50
Kecepatan_angular = 2*math.pi*frekuensi
Sudut = math.pi/6
Periode = 1/frekuensi

def v(t):
    return Vm*np.sin(Kecepatan_angular*t)
def i(t):
    return Im*np.sin(Kecepatan_angular*t-Sudut)
def p(t):
    return v(t)*i(t)

N_plot = 1000               # jumlah titik untuk grafik
t = np.linspace(a, b, N_plot)

v_t = v(t)
i_t = i(t)
p_t = p(t)

plt.figure(figsize=(10, 8))

# Grafik v(t)
plt.subplot(3, 1, 1)
plt.plot(t, v_t)
plt.title('Tegangan v(t)')
plt.ylabel('v(t) [Volt]')
plt.xlabel('t [detik]')
plt.grid(True)

# Grafik i(t)
plt.subplot(3, 1, 2)
plt.plot(t, i_t)
plt.title('Arus i(t)')
plt.ylabel('i(t) [Ampere]')
plt.xlabel('t [detik]')
plt.grid(True)

# Grafik p(t) = v(t) * i(t)
plt.subplot(3, 1, 3)
plt.plot(t, p_t)
plt.title('Daya sesaat p(t) = v(t)  i(t)')
plt.xlabel('t [detik]')
plt.ylabel('p(t) [Watt]')
plt.grid(True)

plt.tight_layout()
plt.show()

```
